## feature selection/ importance using Risk ratio and difference 

```python
for c in categorical:
    print(c)
    data_group=data_full_train.groupby(c).churn.agg(["mean","count"])
    data_group['diff']=data_group["mean"] - global_churn
    data_group["risk"]=data_group["mean"]/global_churn
    display(data_group)
```

#### from here we can see that people with international plan are more likely to churn and people with voice mail plan are less likely to churn and also, state and area code features has low risk ratio and diffrerence score respectively, so its safe to drop these features

## Feature importance mutual information

```python
from sklearn.metrics import mutual_info_score
```

```python
def mutual_info(series):
    return mutual_info_score(series,data_full_train.churn)
```

```python
m_info=data_full_train[categorical].apply(mutual_info)
m_info.sort_values(ascending=False)
```

##### The categorical columns has low mutual scores it is safe to drop some of  them

```python
m_info=data_full_train[numeric].apply(mutual_info)
m_info.sort_values(ascending=False)
```

### one hot encoding with dict vectoriser

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dicts=data_train[categorical+numeric].to_dict(orient='record')
train_dicts[0]
```

```python
dv=DictVectorizer(sparse=False)
dv.fit(train_dicts)
X_train=dv.transform(train_dicts)
dv.get_feature_names()
```

```python
X_train.shape
```

```python
val_dict=data_val[categorical+numeric].to_dict(orient='record')
val_dict[0]
```

```python
X_val=dv.transform(val_dict)
X_val.shape
```

### Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
pred_1 = model.predict_proba(X_val)[:,1]
pred_1
```

```python
churn_decision=(pred_1 >= 0.5)
```
